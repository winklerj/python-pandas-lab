```python
import omniture

#Read credentials file
#Example file format: username|pass-key|https://api-sbx1.omniture.com/admin/1.4/rest/
with open('credentials.txt') as f:
  credentials = [x.strip().split('|') for x in f.readlines()]

#Loop through lines in the file
for username,password,url in credentials:
    #If url is provided
    if url:
        print(username)
        print(password)
        print(url)
        an = omniture.authenticate(username,password,url)
    #If no url is provided
    else:
        print(username)
        print(password)
        an = omniture.authenticate(username,password)

%matplotlib inline
```

```python
suite = an.suites['genclone2.19-23']
report = suite.report.element('page', top=50).metric('pageviews').metric('visits').range('2013-04-01','2013-10-31').run()
```

```python
report.dataframe.plot(kind='scatter',x='pageviews',y='visits')
```

```python
report.dataframe['pv/visit'] = report.dataframe['pageviews']/report.dataframe['visits']
```

```python
report.dataframe.sort('pv/visit', ascending=False)
```
