```python
import omniture
import pandas as pd
import numpy as np
import statsmodels.api as sm

#Read file with credentials
with open('credentials.txt') as f:
  credentials = [x.strip().split('|') for x in f.readlines()]

#Loop through lines in the file
for username,password,url in credentials:
    #If url is provided
    if url:
        print(username)
        print(password)
        print(url)
        an = omniture.authenticate(username,password,url)
    #If no url is provided
    else:
        print(username)
        print(password)
        an = omniture.authenticate(username,password)

%matplotlib inline
```

```python
suite = an.suites['genclone2.19-23']
report = suite.report.metric('pageviews').range('2013-04-01','2013-10-31').granularity("day").run()
```

```python
report.dataframe.plot()
```

```python
arma_mod = sm.tsa.ARMA(report.dataframe['pageviews'],(6,4), dates=report.dataframe['datetime']).fit()
```

```python
#fig = arma_mod.plot_predict('2013-07-01','2013-10-31')
fig = arma_mod.plot_predict('2013-10-31','2013-12-31')
```
